```python
import torch
import matplotlib.pyplot as plt
```

# RNN

```python
train_loss_rnn = torch.load("/home/careinfolab/Dr_Luo/Rohan/ICD_Codes/Results/RNN/train_loss.pth")
train_accuracy_rnn = torch.load("/home/careinfolab/Dr_Luo/Rohan/ICD_Codes/Results/RNN/train_accuracy.pth")
train_f1score_rnn = torch.load("/home/careinfolab/Dr_Luo/Rohan/ICD_Codes/Results/RNN/train_f1score.pth")
train_precision_rnn = torch.load("/home/careinfolab/Dr_Luo/Rohan/ICD_Codes/Results/RNN/train_precision.pth")
train_recall_rnn = torch.load("/home/careinfolab/Dr_Luo/Rohan/ICD_Codes/Results/RNN/train_recall.pth")

test_loss_rnn = torch.load("/home/careinfolab/Dr_Luo/Rohan/ICD_Codes/Results/RNN/test_loss.pth")
test_accuracy_rnn = torch.load('/home/careinfolab/Dr_Luo/Rohan/ICD_Codes/Results/RNN/test_accuracy.pth')
test_f1score_rnn = torch.load("/home/careinfolab/Dr_Luo/Rohan/ICD_Codes/Results/RNN/test_f1score.pth")
test_precison_rnn = torch.load("/home/careinfolab/Dr_Luo/Rohan/ICD_Codes/Results/RNN/test_precision.pth")
test_recall_rnn = torch.load("/home/careinfolab/Dr_Luo/Rohan/ICD_Codes/Results/RNN/test_recall.pth")
```

```python
fig, axs = plt.subplots(2, 3, figsize=(15, 10))

axs[0, 0].plot(range(len(train_accuracy_rnn)), train_accuracy_rnn, label='Train', color='blue')
axs[0, 0].set_title("Accuracy of RNN (from Description to ICD Code)")
axs[0, 0].set_ylabel("Accuracy")
axs[0, 0].legend()

axs[0, 1].plot(range(len(train_precision_rnn)), train_precision_rnn, label='Train', color='blue')
axs[0, 1].set_title("Precision of RNN (from Desciption to ICD Code)")
axs[0, 1].set_ylabel("Precision")
axs[0, 1].legend()

axs[0, 2].plot(range(len(train_recall_rnn)), train_recall_rnn, label='Train', color='blue')
axs[0, 2].set_title("Recall of RNN (from Desc to ICD Code)")
axs[0, 2].set_ylabel("Recall")
axs[0, 2].legend()

axs[1, 0].plot(range(len(train_f1score_rnn)), train_f1score_rnn, label='Train', color='blue')
axs[1, 0].set_title("F1 Score of RNN (from Desc to ICD Code)")
axs[1, 0].set_ylabel("F1")
axs[1, 0].legend()

axs[1, 1].plot(range(len(train_loss_rnn)), train_loss_rnn, label='Train', color='blue')
axs[1, 1].set_title("Loss of RNN (from Desc to ICD Code)")
axs[1, 1].set_ylabel("Loss")
axs[1, 1].legend()

axs[1, 2].axis('off')

plt.tight_layout()
plt.show()
```

```python
fig, axs = plt.subplots(2, 3, figsize=(15, 10))

axs[0, 0].plot(range(len(test_accuracy_rnn)), test_accuracy_rnn, label='Test', color='orange')
axs[0, 0].set_title("Accuracy of RNN (from Description to ICD Code)")
axs[0, 0].set_ylabel("Accuracy")
axs[0, 0].legend()

axs[0, 1].plot(range(len(test_precison_rnn)), test_precison_rnn, label='Test', color='orange')
axs[0, 1].set_title("Precision of RNN (from Description to ICD Code)")
axs[0, 1].set_ylabel("Precision")
axs[0, 1].legend()

axs[0, 2].plot(range(len(test_recall_rnn)), test_recall_rnn, label='Test', color='orange')
axs[0, 2].set_title("Recall of RNN (from Desc to ICD Code)")
axs[0, 2].set_ylabel("Recall")
axs[0, 2].legend()

axs[1, 0].plot(range(len(test_f1score_rnn)), test_f1score_rnn, label='Test', color='orange')
axs[1, 0].set_title("F1 Score of RNN (from Desc to ICD Code)")
axs[1, 0].set_ylabel("F1")
axs[1, 0].legend()

axs[1, 1].plot(range(len(test_loss_rnn)), test_loss_rnn, label='Test', color='orange')
axs[1, 1].set_title("Loss of RNN (from Desc to ICD Code)")
axs[1, 1].set_ylabel("Loss")
axs[1, 1].legend()

axs[1, 2].axis('off')

plt.tight_layout()
plt.show()
```
