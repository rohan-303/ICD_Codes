```python
def add_dot(s):
    return s[:3]+'.'+s[3:]
```

```python
for i in range(len(df)):
    if len(df[0].iloc[i])>3:
        df.loc[i, 4] = df[4].iloc[i] + ', ' + df[5].iloc[i]
        df.loc[i, 2] = add_dot(df[2].iloc[i])
```

```python
df.head(10)
```

```python
df = df.drop([0,1,3,5], axis=1)
df = df.set_axis(['ICD_Code', 'Description'], axis=1)
```

```python
df.head(5)
```

```python
for i in range(len(df)):
    if '.' not in df['ICD_Code'].iloc[i]:
        df.loc[i, 'ICD_Code'] = add_dot(df['ICD_Code'].iloc[i])
```

```python
df.head(10)
```

```python
df1 = df.copy()
```

```python
def clean_text(text):
    text = text.lower()                       
    text = re.sub(r'[^a-z0-9\s]', '', text)   
    text = re.sub(r'\s+', ' ', text).strip()   
    return text
```

```python
df1['Description'] = df1['Description'].apply(clean_text)
```
